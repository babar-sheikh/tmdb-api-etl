```python
import requests
import pandas as pd

url = "https://api.themoviedb.org/3/movie/top_rated"
api_key = "3acfd3043b61811d8de5f205b8bbaa80"
headers = {"accept": "application/json"}

# List to store data from all pages
all_movies = []

# TMDB usually limits you to 500 pages, but we'll use your range (1 to 428)
for i in range(1, 300):
    params = {
        "api_key": api_key,
        "language": "en-US",
        "page": i
    }
    
    try:
        response = requests.get(url, headers=headers, params=params)
        response.raise_for_status()
        
        page_data = response.json().get('results', [])
        
        # Add this page's movies to our master list
        all_movies.extend(page_data)
        
        # Optional: Print progress every 50 pages
        if i % 50 == 0:
            print(f"Fetched page {i}...")
9
    except requests.exceptions.RequestException as e:
        print(f"Error on page {i}: {e}")
        break  # Stop if there's a connection issue

# Create the final DataFrame after the loop finishes
df = pd.DataFrame(all_movies)[['id', 'title', 'overview', 'popularity', 'release_date', 'vote_average', 'vote_count']]

print("Data collection complete!")
print(df.head())
```
